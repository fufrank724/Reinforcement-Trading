```python
import pandas as pd
import numpy as np

input_data=pd.read_csv("TXF_60min_filted.CSV")
input_data.head()
```

```python
input_data.drop('quote', axis=1, inplace=True)
input_data.head()
```

```python
import datetime
date=input_data['Date'][0]
print(type(date))
date=(datetime.datetime.strptime(date,"%Y/%m/%d"))
year=date.strftime('%Y')
print(year)
```

```python
state=[]
current_step=0
def make_state():
    current_step=0
    for i in range(5):
        state.append([input_data['H_L'][current_step],
                     input_data['H_C'][current_step],
                     input_data['C_L'][current_step],
                     input_data['avg_c10_filt'][current_step],
                     input_data['avg_vol10_filt'][current_step]])
        current_step+=1
make_state()
print(current_step)
print("state = ",state)
```
